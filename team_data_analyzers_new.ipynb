```python
corpus = []
#content = file.readlines()
for row in range(len(X_train.index)):
    value1 = X_train.iat[row, 1]
    value2 = X_train.iat[row, 0]
    line1 = content[value1-1]
    line2 = content[value2-1]
    final_line=line1+line2
    review = re.sub('[^a-zA-Z]', ' ', final_line)
    review = review.lower()
    review = review.split()
    #review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
with open(r'./corpus.txt', 'w') as fp:
    for item in corpus:
        # write each item on a new line
        fp.write("%s\n" % item)
```
